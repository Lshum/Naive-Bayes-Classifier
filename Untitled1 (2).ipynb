```python
import sklearn
import pandas as pd
import numpy as np
import re
import decimal
import math
import collections
from collections import Counter
from math import log
from nltk.tokenize import word_tokenize
#Reading the training set file and test file
training_data = pd.read_csv("/Users/Lawrence/Downloads/trg.csv")
test_set = pd.read_csv("/Users/Lawrence/Downloads/tst.csv")
#Finding Unique Words
words = []
for i in range(0, len(training_data)):
    abstract = training_data['abstract'][i]
    word_abstract = abstract.split(" ")
    words.extend(word_abstract)

words = len(list(np.unique(np.sort(words))))

#Sorting all the words into different class labels
A,B,E,V = "","","",""
for row1, row2 in training_data.iterrows():
    a, t = row2['class'], row2['abstract']
    if a == "A":
        A += " " + t
    elif a == "B":
        B += " " + t
    elif a == "E":
        E += " " + t
    elif a == "V":
        V += " " + t

#Number of instances
length = len(training_data)
#Calculating the probability of each class
classes = dict(Counter(training_data['class']))
for a,b in classes.items():
    classes[a] = b/length

count_A = Counter(word_tokenize(A))
count_B = Counter(word_tokenize(B))
count_E = Counter(word_tokenize(E))
count_V = Counter(word_tokenize(V))

#Calculating the conditional probability of each word appearing in a class label
def prob(data):
    for a,b in data.items():
        total = sum(data.values())
        outcome = float((b+1))/float((total + words))
        data[a] = (outcome)
    return data

class_A = prob(count_A)
class_B = prob(count_B)
class_E = prob(count_E)
class_V = prob(count_V)


test = test_set['abstract']
prior1 = log(classes['B'])
prior2 = log(classes['A'])
prior3 = log(classes['E'])
prior4 = log(classes['V'])
table1 = []
table2 = []
table3 = []
table4 = []

#Running through each word in the test set and calculating the probabiity of each word occuring and adding another else statement if a word does not appear in the dictionary
for i in range(0, len(test)):
    data = test[i].split(" ")
    result1 = 1
    for j in range(0, len(data)):
        if data[j] in class_B.keys():
            result1 += log(class_B.get(data[j]))
        else:
            result1 += 20
    result1 = result1 + prior1
    table1.append(result1)

for i in range(0, len(test)):
    data = test[i].split(" ")
    result2 = 1
    for j in range(0, len(data)):
        if data[j] in class_A.keys():
            result2 += log(class_A.get(data[j]))
        else:
            result2 += 20
    result2 = result2 + prior2
    table2.append(result2)
        
for i in range(0, len(test)):
    data = test[i].split(" ")
    result3 = 1
    for j in range(0, len(data)):
        if data[j] in class_E.keys():
            result3 += log(class_E.get(data[j]))
        else:
            result3 += 20
    result3 = result3 + prior3
    table3.append(result3)
        
for i in range(0, len(test)):
    data = test[i].split(" ")
    result4 = 1
    for j in range(0, len(data)):
        if data[j] in class_V.keys():
            result4 += log(class_V.get(data[j]))
        else:
            result4 += 20
    result4 = result4 + prior4
    table4.append(result4)

#Creating a dataframe of each class label and comparing the probability and choose the minimum value as it is in logarithms.
df = pd.DataFrame()
df["A"] = table2
df["B"] = table1
df["E"] = table3
df["V"] = table4
df['Max'] = df.idxmin(axis = 1)
new_list = df['Max'].tolist()
new_list

#Creating a csv file with the class label and the description for kaggle
test_set_class_predictions = new_list
test_set["class"] = test_set_class_predictions
test_set.drop(["abstract"], axis = 1).to_csv("tst_kaggle.csv", index=False)
```
